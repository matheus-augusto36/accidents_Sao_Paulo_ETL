```python
colunas_veiculo = [
    'tp_veiculo_bicicleta', 'tp_veiculo_caminhao', 'tp_veiculo_motocicleta',
    'tp_veiculo_nao_disponivel', 'tp_veiculo_onibus', 'tp_veiculo_outros',
    'tp_veiculo_automovel'
]


df_sinistros = df_sinistros.melt(
    id_vars=[col for col in df_sinistros.columns if col not in colunas_veiculo],  
    value_vars=colunas_veiculo,
    var_name='tipo_de_veiculo',
    value_name='valor'
)
```

```python
df_sinistros = df_sinistros.dropna(subset=['valor']).drop(columns='valor')
```

```python
df_sinistros['tipo_de_veiculo'] = df_sinistros['tipo_de_veiculo'].str.replace('tp_veiculo_', '')
```

```python
df_sinistros.info()
```

```python
colunas_gravidade = [
    'gravidade_nao_disponivel', 'gravidade_leve', 'gravidade_fatal',
    'gravidade_ileso', 'gravidade_grave'
]


df_sinistros = df_sinistros.melt(
    id_vars=[col for col in df_sinistros.columns if col not in colunas_gravidade],  
    value_vars=colunas_gravidade,
    var_name='gravidade',
    value_name='valor'
)
```

```python
df_sinistros = df_sinistros.dropna(subset=['valor']).drop(columns='valor')
```

```python
df_sinistros['gravidade'] = df_sinistros['gravidade'].str.replace('gravidade_', '')
```

```python
df_sinistros.info()
```

```python
colunas_colisao = [
    'tp_sinistro_atropelamento', 'tp_sinistro_colisao_frontal', 'tp_sinistro_colisao_traseira',
    'tp_sinistro_colisao_lateral', 'tp_sinistro_colisao_transversal', 'tp_sinistro_colisao_outros',
    'tp_sinistro_choque', 'tp_sinistro_capotamento', 'tp_sinistro_engavetamento', 'tp_sinistro_tombamento', 'tp_sinistro_outros',
    'tp_sinistro_nao_disponivel' 
]


df_sinistros = df_sinistros.melt(
    id_vars=[col for col in df_sinistros.columns if col not in colunas_colisao],  
    value_vars=colunas_colisao,
    var_name='tipo_sinistro',
    value_name='valor'
)
```

```python
df_sinistros = df_sinistros.dropna(subset=['valor']).drop(columns='valor')
```

```python
df_sinistros['tipo_sinistro'] = df_sinistros['tipo_sinistro'].str.replace('tp_sinistro_', '')
```

```python
df_sinistros.tipo_de_veiculo.value_counts()
```

```python
df_sinistros['tipo_de_veiculo'] = df_sinistros.tipo_de_veiculo.replace({
    'automovel': 'Automóvel',
    'motocicleta': 'Motocicleta',
    'caminhao': 'Caminhão',
    'onibus': 'Ônibus',
    'outros': 'Outros',
    'bicicleta': 'Bicicleta'
})
```

```python
df_sinistros['tipo_sinistro'] = df_sinistros['tipo_sinistro'].replace({
    'colisao_lateral': 'Colisão Lateral',
    'outros': 'Outros',
    'colisao_frontal': 'Colisão Frontal',
    'colisao_traseira': 'Colisão Traseira',
    'choque': 'Choque',
    'tombamento': 'Tombamento',
    'atropelamento': 'Atropelamento',
    'colisao_transversal': 'Colisão Transversal',
    'capotamento': 'Capotamento',
    'engavetamento': 'Engavetamento',
    'colisao_outros': 'Outros'
})
```

```python
df_sinistros['gravidade'] = df_sinistros['gravidade'].replace({
    'leve': 'Leve',
    'grave': 'Grave',
    'fatal': 'Fatal'
})
```
