```python
import matplotlib.pyplot as plt

def plotar_nulos(df, titulo="Percentual de Nulos por Coluna"):
    percentual = (df.isnull().sum() / len(df) * 100).round(2)
    percentual = percentual[percentual > 0]  # exibir apenas colunas com nulos

    if percentual.empty:
        print("✅ Nenhuma coluna com valores nulos.")
        return

    plt.figure(figsize=(10, 5))
    bars = percentual.sort_values(ascending=False).plot(kind='bar', color='orange', edgecolor='black')

    plt.title(titulo)
    plt.ylabel('Percentual (%)')
    plt.xlabel('Colunas')
    plt.xticks(rotation=45, ha='right')
    plt.ylim(0, 100)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()

    # Rótulos de dados no topo das barras
    for idx, value in enumerate(percentual.sort_values(ascending=False)):
        plt.text(idx, value + 1, f'{value:.1f}%', ha='center', va='bottom', fontsize=9)

    plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def analisar_nulos_por_grupo(df, coluna_referencia, top_n=10):
    """
    Para cada coluna com valores nulos, calcula e plota o percentual de nulos por grupo,
    baseado na coluna de referência (ex: 'tipo_de_vitima').
    
    Parâmetros:
    - df: DataFrame
    - coluna_referencia: coluna categórica para agrupar os dados (ex: tipo de vítima)
    - top_n: número máximo de categorias para exibir no gráfico
    
    """
    colunas_com_nulos = df.columns[df.isna().any()]
    
    for coluna in colunas_com_nulos:
        # Agrupar e calcular percentual de nulos por grupo
        nulos_por_grupo = (
            df[df[coluna].isna()]
            .groupby(coluna_referencia)
            .size()
            / df.groupby(coluna_referencia).size()
        ) * 100

        nulos_por_grupo = nulos_por_grupo.dropna().sort_values(ascending=False)[:top_n]

        # Exibir o resultado
        print(f"\nColuna analisada: {coluna}")
        print(nulos_por_grupo.round(2))

        # Plotar o gráfico
        plt.figure(figsize=(10, 4))
        sns.barplot(x=nulos_por_grupo.index, y=nulos_por_grupo.values)
        plt.title(f"% de valores nulos em '{coluna}' por '{coluna_referencia}'")
        plt.ylabel('% de nulos')
        plt.xlabel(coluna_referencia)
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
```

```python
analisar_nulos_por_grupo(df_fatais, coluna_referencia='profissao')
```
