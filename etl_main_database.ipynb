```python
import matplotlib.pyplot as plt

def plotar_nulos(df, titulo="Percentual de Nulos por Coluna"):
    percentual = (df.isnull().sum() / len(df) * 100).round(2)
    percentual = percentual[percentual > 0]

    if percentual.empty:
        print("✅ Nenhuma coluna com valores nulos.")
        return

    plt.figure(figsize=(10, 5))
    bars = percentual.sort_values(ascending=False).plot(kind='bar', color='orange', edgecolor='black')

    plt.title(titulo)
    plt.ylabel('Percentual (%)')
    plt.xlabel('Colunas')
    plt.xticks(rotation=45, ha='right')
    plt.ylim(0, 100)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()

    # Rótulos de dados no topo das barras
    for idx, value in enumerate(percentual.sort_values(ascending=False)):
        plt.text(idx, value + 1, f'{value:.1f}%', ha='center', va='bottom', fontsize=9)

    plt.show()
```

```python
# Análise de outliers

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))
sns.boxplot(x=df_acidentes['idade'])
plt.title(f'Boxplot de idade - Identificação de Outliers')
plt.xlabel('idade')
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 4))
sns.boxplot(x=df_acidentes['ano_fab'])
plt.title(f'Boxplot de ano_fab - Identificação de Outliers')
plt.xlabel('ano_fab')
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 4))
sns.boxplot(x=df_acidentes['ano_modelo'])
plt.title(f'Boxplot de ano_modelo - Identificação de Outliers')
plt.xlabel('ano_modelo')
plt.grid(True)
plt.show()
```

```python
# Remoção de outliers

df_acidentes = df_acidentes[(df_acidentes['idade'] >= 0) & (df_acidentes['idade'] <= 120)]
```

```python
df_acidentes['ano_fab'] = df_acidentes['ano_fab'].replace(0.0, None)
df_acidentes['ano_modelo'] = df_acidentes['ano_modelo'].replace(0.0, None)
```

```python
# Remoção de outliers

df_acidentes = df_acidentes[((df_acidentes['ano_fab'] >= 1900) & (df_acidentes['ano_fab'] <= 2025)) & ((df_acidentes['ano_modelo'] >= 1900) & (df_acidentes['ano_modelo'] <= 2025))]
```
